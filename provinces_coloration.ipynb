### RESOLUTION: 
* ISA + ROA = TELO 

```python
# Variables
input_1 = "ISA"
input_2 = "ROA"
input_3 = "TELO"

# Domains
domains = range(10)

# Contraintes 
""" input_1 + input_2 = input_3 """

letters_pb_1 = set(input_1 + input_2 + input_3)
letters_pb_1
```

```python
def input_to_numbers(input, solution):
    result = 0
    for index, letter in enumerate(input):
        result += solution[letter] * (10 ** (len(input) - (index + 1)))
        
    return result
```

```python
def resolve(input_1, input_2, input_3, domains, letters):
    for perm in permutations(domains, len(letters)):
        solution = dict(zip(letters, perm))
        
        if len(input_3) > len(input_2) and len(input_3) > len(input_1):
            if solution.get(input_3[0]) != 1:
                    continue
            
        first, second, third = 0, 0, 0
        
        first = input_to_numbers(input_1, solution)
        second = input_to_numbers(input_2, solution)
        third = input_to_numbers(input_3, solution)
        
        if first + second == third:
            return first, second, third, solution
                
```

```python
def display_solution(input_1, input_2, input_3, domains, letters):
    first, second, third, solution = resolve(input_1, input_2, input_3, domains, letters)
    print(f"{input_1} + {input_2} = {input_3}")
    print(f"{first} + {second} = {third}")
    print(f"Solution: {solution}")
```

* SEND + MORE = MONEY

```python
# Variables 
send = "SEND"
more = "MORE"
money = "MONEY"

letters_pb_2 = set(send + more + money)
letters_pb_2
```

```python
display_solution(input_1, input_2, input_3, domains, letters_pb_1)
display_solution(send, more, money, domains, letters_pb_2)
```
